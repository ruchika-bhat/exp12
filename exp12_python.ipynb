```python
# second dataset
import pandas as pd

df = pd.read_csv("/content/Expt11 (1).csv")
df
```

```python
# find how many students obtained each grade

print(df['Grade'].value_counts())
```

```python
# find how many males and females are there

print(df['Gender'].value_counts())
```

```python
# find how many students are in each department

print(df['Department'].value_counts())
```

```python
# percentage distribution

#(normalize=True) this command helps it to divide with total number
print(df['Grade'].value_counts(normalize=True)*100)
print(df['Gender'].value_counts(normalize=True)*100)
print(df['Department'].value_counts(normalize=True)*100)
```

```python
# gender vs grade analysis

print(pd.crosstab(df['Gender'],df['Grade']))
```

```python
print(pd.crosstab(df['Department'],df['Gender']))
```

```python
print(pd.crosstab(df['Department'],df['Grade']))
```

```python
# percentage analysis

print(pd.crosstab(df['Department'],df['Grade'],normalize='index')*100)
# we use index to divide with row ka total
```

```python
# grouping data

print(df.groupby('Department')['Grade'].value_counts())
```

```python
# new data
import pandas as pd

data = {
'Order_ID':['O1','O2','O3','O4','O5','O6','O7','O8','O9','O10'],

'Category':['Electronics','Clothing','Electronics','Grocery','Clothing',
            'Electronics','Grocery','Clothing','Electronics','Grocery'],

'Payment_Method':['UPI','Card','UPI','Cash','Card',
                  'UPI','Cash','Card','UPI','Cash'],

'Delivery_Type':['Express','Standard','Express','Standard','Express',
                 'Standard','Express','Standard','Express','Standard'],

'Customer_Type':['New','Returning','Returning','New','Returning',
                 'New','Returning','New','Returning','New']
}
df = pd.DataFrame(data)
df
```

```python
# count how many orders in each category

print(df['Category'].value_counts())
```

```python
# percentage distribution

print(df['Category'].value_counts(normalize=True)*100)
```

```python
print(df['Payment_Method'].value_counts())
```

```python
# display all unique categories

print(df['Category'].unique())
```

```python
# no. of unique values

print(df['Category'].nunique())
```

```python
print(pd.crosstab(df['Category'],df['Payment_Method']))
```

```python
# filtering data
# example: orders from electronic category

electonics_orders = df[df['Category']=='Electronics']
print(electonics_orders)
```

```python
# sorting alphabetically

sorted_df = df.sort_values(by='Category')
print(sorted_df)
```

```python
print(df.groupby('Category')['Payment_Method'].value_counts())
```
